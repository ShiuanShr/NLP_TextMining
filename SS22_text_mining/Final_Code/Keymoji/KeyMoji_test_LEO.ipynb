```python
def store_to_csv(data, path):
  with open(path, 'w', newline='') as csvfile:
      writer = csv.writer(csvfile)
      writer.writerow(data)

def store_to_csv_df(data, path):
  with open(path, 'w', encoding = 'utf-8-sig') as f:
    data.to_csv(f)
def clean(string_content):
  article_1 = [article.replace('\n', '') for article in string_content]
#  [、_ =\ \\ [\ ] ; ' ，. / ~ ! @ # $ % ^ & *  + | ? > < 「 :{}]

  final_list = [re.sub('[^\u4E00-\u9FA50-9 + ]', '', i) for i in article_1] # ，! @ # ? & ? > < 「」」「 {}
  final_list = [re.sub(' ', '', i) for i in final_list]

  return final_list
```

```python
import time # 引入time


def gettime():
  nowTime = int(time.time()) # 取得現在時間
  struct_time = time.localtime(nowTime) # 轉換成時間元組
  rr = time.strftime("%Y %m %d %I:%M:%S %P", struct_time) # 將時間元組轉換成想要的字串
  return rr
gettime()
```

```python
# from KeyMojiAPI import KeyMoji


def kmo(dataset):
  save = 0
  iterated_df = dataset.copy()
  # # 若您是使用 Docker 版本，無須填入 username, keymoji_key 參數
  keymoji = KeyMoji(username="", keymojiKey="")
  restart = False
  KeyMoji_score = []
  
  save_path_Keymoji = "/content/drive/Shareddrives/Text Mining/CSV/temp_leo1/"

  for j in range(10): # 每次最多跑10次
    if restart== False:
      for i in range(len(iterated_df)): # 每次最多跑100筆
        print('4')
        i_dict = keymoji.sense8(iterated_df[0][i])
        print('5')
        if i_dict.get("msg") == 'Success!':
          print(f" The sentence {i} is completed!")
          print(i_dict)
          KeyMoji_score.append(i_dict)
          temp0 = KeyMoji_score
          print('55')
        else:
          print(f" Coda is running out in {i} stage; recall the function again")
          print('5-1')
          save+=1
          print('5-2')
          save_order(temp0,save_path_Keymoji,save)
          print('5-4')
          restart = True
          temp = iterated_df[i:]
          print(gettime())
          break
    else:
      print('i an here')
      iterated_df = temp
      restart = False

def save_order (score, path,save):
  dff = pd.DataFrame(score)
  dff.columns = ['Title', 'Anger', 'Anticipation', 'Disgust', 'Fear', 'Joy', 'Sadness', 'Surprise', 'Trust']
  name = "KeyMoji_score_ALL_TAI" + f"_{save}"+".csv"
  print('dfdd')
  path = path + name
  with open(path, 'w', encoding = 'utf-8-sig') as f:
    dff.to_csv(f)
  print(f'Done: save {save}')
```
