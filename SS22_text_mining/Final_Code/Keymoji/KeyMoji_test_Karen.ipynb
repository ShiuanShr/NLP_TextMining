```python
def clean(string_content):
  article_1 = [article.replace('\n', '') for article in string_content]
#  [、_ =\ \\ [\ ] ; ' ，. / ~ ! @ # $ % ^ & *  + | ? > < 「 :{}]

  final_list = [re.sub('[^\u4E00-\u9FA50-9 + ]', '', i) for i in article_1] # ，! @ # ? & ? > < 「」」「 {}
  final_list = [re.sub(' ', '', i) for i in final_list]

  return final_list
```

```python
def store_to_csv(data, path):
  with open(path, 'w', newline='') as csvfile:
      writer = csv.writer(csvfile)
      writer.writerow(data)
```

```python
def store_to_csv_df(data, path):
  with open(path, 'w', encoding = 'utf-8-sig') as f:
    data.to_csv(f)
```

```python
x = []
KeyMoji_result = []
# 在range 中填入成功index的range
for i in range(0,17):
  x = np.array([KeyMoji_score[i]['results'][0]['input_str'],
          KeyMoji_score[i]['results'][0]['Anger'],
          KeyMoji_score[i]['results'][0]['Anticipation'],
          KeyMoji_score[i]['results'][0]['Disgust'],
          KeyMoji_score[i]['results'][0]['Fear'],
          KeyMoji_score[i]['results'][0]['Joy'],
          KeyMoji_score[i]['results'][0]['Sadness'],
          KeyMoji_score[i]['results'][0]['Surprise'],
          KeyMoji_score[i]['results'][0]['Trust']])
  KeyMoji_result.append(x)
```

```python
# 存入CSV
YANG_KeyMoji_score = pd.DataFrame(KeyMoji_result)
YANG_KeyMoji_score.columns = ['Title', 'Anger', 'Anticipation', 'Disgust', 'Fear', 'Joy', 'Sadness', 'Surprise', 'Trust']
name = "KeyMoji_score_ALL_YANG.csv"

path = save_path_Keymoji + name
store_to_csv_df(YANG_KeyMoji_score, path)
```
