```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from PIL import Image #???
import jieba
import numpy as np
print(jieba.__file__)
# 如果檔案內有一些編碼錯誤，使用 errors='ignore' 來忽略錯誤
with open("../data/詩經.txt", encoding="big5", errors='ignore') as f:
    text = f.read()

# 設定使用 big5 斷詞
jieba.set_dictionary('../data/dict.txt.big.txt') 
#此用法要下載big5斷詞檔案
##不一定要設定詞庫，內建的效果也不錯


wordlist = jieba.cut(text)
print('wordlist:',type(wordlist))

words = " ".join(wordlist)
print(type(words))
#文字雲造型圖片
mask = np.array(Image.open('../data/parrot-by-jose-mari-gimenez2.jpg')) #文字雲形狀
# 從 Google 下載的中文字型
font = 'simsun.ttc'  #微軟宋黑體 若error 無法開啟resource 則可能C Disk內沒有灌

#背景顏色預設黑色，改為白色、使用指定圖形、使用指定字體
my_wordcloud = WordCloud(background_color='white',mask=mask,font_path=font).generate(words)
#   # background_color设备背景颜色,                   
#font_path要设置，不然会有乱码出现。

plt.axis("off")

plt.imshow(my_wordcloud, interpolation = 'nearest')

#存檔
my_wordcloud.to_file('word_cloud.png')
```

# 自定義背景圖 進行英文文字雲

#基本觀念便是，將 wordcloud 產生的圖片生成為透明背景、
#然後使用 Python 中的 PIL 模組將其與背景圖片 composite 在一起即可。

reference:
https://clay-atlas.com/blog/2020/04/24/python-cn-package-wordcloud-custom-background-image/

```python
import matplotlib.pyplot as plt
from PIL import Image 
#基本觀念便是，將 wordcloud 產生的圖片生成為透明背景、
#然後使用 Python 中的 PIL 模組將其與背景圖片 composite 在一起即可。
from wordcloud import WordCloud 


# Load 文字text檔
text = open('data/t8.shakespeare.txt', 'r', encoding='utf-8').read()


# WordCloud 產生背景
wc = WordCloud(mode='RGBA', background_color='rgba(255, 255, 255, 0)')
wc.generate(text)  


# Image process
image = Image.fromarray(wc.to_array())

#image = image.convert("P") #將image轉成P mode 
background = Image.open('data/image-46.png') #mode可能不是RGBA image has wrong mode
print('background.mode: ',background.mode)
print('background.size: ',background.size)
print('image.mode: ',image.mode)
print('image.size: ',image.size)

background = background.resize(image.size) #使兩張背景size相同 (400, 200)
print('background.size after resize: ',background.size)
background = background.convert("RGBA") #使兩個都是RGBA
print('background.size after mode change: ',background.mode)

new_image = Image.alpha_composite(background, image)
# combine 2 image (the internal built background and the specific background )
# both of them should be RGBA

# Plot
plt.figure()
plt.axis('off')
plt.imshow(new_image)
plt.show()


# Save
plt.figure()
plt.axis('off')
fig = plt.imshow(new_image, interpolation='nearest')
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html

fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
plt.savefig('test.png',
            bbox_inches='tight',
            pad_inches=0,
            format='png',
            dpi=300)
```
