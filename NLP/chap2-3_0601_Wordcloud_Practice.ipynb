```python
import matplotlib.pyplot as plt
from PIL import Image 
# wordcloud 
# Python  PIL  composite 
from wordcloud import WordCloud 


# Load text
text = open('data/t8.shakespeare.txt', 'r', encoding='utf-8').read()


# WordCloud 
wc = WordCloud(mode='RGBA', background_color='rgba(255, 255, 255, 0)')
wc.generate(text)  


# Image process
image = Image.fromarray(wc.to_array())

#image = image.convert("P") #imageP mode 
background = Image.open('data/image-46.png') #modeRGBA image has wrong mode
print('background.mode: ',background.mode)
print('background.size: ',background.size)
print('image.mode: ',image.mode)
print('image.size: ',image.size)

background = background.resize(image.size) #size (400, 200)
print('background.size after resize: ',background.size)
background = background.convert("RGBA") #RGBA
print('background.size after mode change: ',background.mode)

new_image = Image.alpha_composite(background, image)
# combine 2 image (the internal built background and the specific background )
# both of them should be RGBA

# Plot
plt.figure()
plt.axis('off')
plt.imshow(new_image)
plt.show()


# Save
plt.figure()
plt.axis('off')
fig = plt.imshow(new_image, interpolation='nearest')
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html

fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
plt.savefig('test.png',
            bbox_inches='tight',
            pad_inches=0,
            format='png',
            dpi=300)
```

```python
def calculate_frequencies(file_contents):
    # Here is a list of punctuations and uninteresting words you can use to process your text
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''  #unnecessary stuffs
    uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
    "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just"]
    
    # LEARNER CODE START HERE
    non_punctuation="" 
    #First: eliminate the punctuations
    for char in file_contents:
        if char not in punctuations:
            non_punctuation=non_punctuation + char
            
    words_list= non_punctuation.split()
    #print('words:{}'.format(words_list))
    clean_words=[]  #to load the result of isalpha from words's split
    frequencies={}  

    for word in words_list:  # alphabets OR NOT
        if word.isalpha():
            if word not in uninteresting_words: #ignore uninteresting or irrelevant words
                clean_words.append(word)

    for alphet in clean_words: 
        if alphet.lower() not in frequencies: #ignore case
            frequencies[alphet.lower()]=1     #count the frequencies

        else:
            frequencies[alphet.lower()]+=1
            
    #wordcloud
    cloud = wordcloud.WordCloud() #sometimes error likes 'wordcloud have no attribute wordcloud.' Solution: restart the backend sever or rename the file in your notebook
    cloud.generate_from_frequencies(frequencies)
    return cloud.to_array()

file_contents='2020 came and went fast, but fortunately, the publishing industry kept pace with the passage of time with a slew of the years most anticipated titles. Here, take a look back at the best new books that arrived this yearand add them to your 2021 reading list if you haven dug into them yet.From a black, queer writer and former biochem Ph.D. candidate living in a Midwestern university town comes a searing debut about  a black, queer biochem Ph.D. candidate living in a Midwestern university town. When Wallace has an unexpected encounter with a supposedly-straight white classmate amid a time of mounting hostility in his community, he is forced to confront long-hidden wounds. Whether despite or because of Taylors closeness to his subject matter, the result is a novel of quiet, startling power.Ever since the publication of Meaty in 2013, Irbys essays have been required reading on the millennial condition. In her latest collection, the writernow approaching 40 and living a Pinterest-ified version of the American dream in a small Midwestern townturns her addictively bummed-out wit to topics like lesbian bed death and the difficulty of making adult friendships.' 
myimage = calculate_frequencies(file_contents)
plt.imshow(myimage, interpolation = 'nearest')
plt.axis('off')
plt.show()
```
