## 使用 sciki-learn 套件計算 TF-IDF(英文文本)
    

### 導入套件

```
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

### 讀取資料集

```
docs = [
    'John likes to watch movies. Mary likes movies too.',
    'John also likes to watch football games.',
    'Mary also likes to play video games.'
]
```

```
vec = TfidfVectorizer()
matrix =  vec.fit_transform(docs)
print(matrix)

tfidfs_list = matrix.toarray()

print(tfidfs_list)

tokens = vec.get_feature_names()
pd.DataFrame(tfidfs_list, columns = tokens)
```

### 計算 tf-idf

```
vectorizer = TfidfVectorizer()            #類似tf的用法
document_term_matrix = vectorizer.fit_transform(docs) #train 固定用法
tfidfs = document_term_matrix.toarray()        #matrix 轉array
tokens = vectorizer.get_feature_names()   # 3 *12的陣列
pd.DataFrame(tfidfs,columns=tokens)
```

### 查看 idf

```
pd.DataFrame([vectorizer.idf_],columns=tokens)

#不懂??? 這樣我是哪個文件的?
```

```


titles = df['title'].apply(str).tolist() #取出column為title的欄位轉str再tolist


contents = df['content'].apply(str).tolist()

selected_docs = [2,3,11]

docs = []
for i in selected_docs:
    docs.append(contents[i])
    print(contents[i])
```

```
aa = df['title'][:3]
print('aa: \n',aa)
print(type(aa))

print('\n',('*'*20))

bb = df['title'][:3].apply(str) 
print('bb: \n',bb)

print('\n',('*'*20))
print(type(bb))

cc = bb.tolist()
print('cc: \n',cc)

print('\n',('*'*20))
print(type(cc))

dd = aa.tolist()
print('dd: \n',dd)
print(type(dd))

#結論 寫apply(str)沒意義
```

```
# 分詞再合併
import jieba

tokenized_docs = []
for doc in docs:
    tokens = jieba.lcut(doc)
    doc = ' '.join(tokens)
    tokenized_docs.append(doc)
```

```
# 計算tf-idf

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
document_term_matrix = vectorizer.fit_transform(tokenized_docs)
tfidfs = document_term_matrix.toarray()
tokens = vectorizer.get_feature_names()
pd.DataFrame(tfidfs,columns=tokens)
```

```
# tfidf最大值
max([max(doc) for doc in tfidfs])


```

```
#因文字雲不吃float 只吃int 需要進一步修改
import numpy as np

a = [1,2,3]
b = np.array(a)
# map 功能: 
# 類似pd.apply 可針對可迭代的變數中每一元素套用來執行特定function
# 
x = list(map(float,a)) #a 轉float
y = np.array(list(map(float,b))) #b 轉flaot

print(x)
print(y)
```

```
# 將 tf-idf 小數轉換成整數

#方法一：
int_tfidfs = [list(map(int,tfidf*100)) for tfidf in tfidfs] 
print(int_tfidfs[0])

#方法二：
int_tfidfs = [[int(tfidf*100) for tfidf in doc] for doc in tfidfs] #從又看到左來看迴圈 = 由內至外
print(int_tfidfs[0])
```

```
# 依三個文件個別產生文字雲

wcs = []
for tfidf in int_tfidfs:
    wc = WordCloud(background_color='white',
               font_path='/content/taipei_sans_tc_beta.ttf')
    ct = dict(zip(tokens,tfidf))
    ct = {x:y for x,y in ct.items() if y!=0}
    wc.generate_from_frequencies(ct)
    wcs.append(wc)
```
